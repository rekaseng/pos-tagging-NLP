```python
#TF_IDF 
import numpy as np
from nltk.tokenize import  word_tokenize 
import random
import string
import bs4 as bs
import re
```

```python
import nltk
nltk.download('punkt')
```

```python
corpuss = nltk.sent_tokenize(corpus)

for i in range(len(corpuss )):
    corpuss [i] = corpuss [i].lower()
    corpuss [i] = re.sub(r'\W',' ',corpuss [i])
    corpuss [i] = re.sub(r'\s+',' ',corpuss [i])

wordfreq = {}
for sentence in corpuss:
    tokens = nltk.word_tokenize(sentence)
    for token in tokens:
        if token not in wordfreq.keys():
            wordfreq[token] = 1
        else:
            wordfreq[token] += 1

import heapq
most_freq = heapq.nlargest(10, wordfreq, key=wordfreq.get)
```

```python
word_idf_values = {}
for token in most_freq:
    doc_containing_word = 0
    for document in corpuss:
        if token in nltk.word_tokenize(document):
            doc_containing_word += 1
    word_idf_values[token] = np.log(len(corpus)/(1 + doc_containing_word))
```

```python
word_tf_values = {}
for token in most_freq:
    sent_tf_vector = []
    for document in corpuss:
        doc_freq = 0
        for word in nltk.word_tokenize(document):
            if token == word:
                  doc_freq += 1
        word_tf = doc_freq/len(nltk.word_tokenize(document))
        sent_tf_vector.append(word_tf)
    word_tf_values[token] = sent_tf_vector
```

```python
tfidf_values = []
for token in word_tf_values.keys():
    tfidf_sentences = []
    for tf_sentence in word_tf_values[token]:
        tf_idf_score = tf_sentence * word_idf_values[token]
        tfidf_sentences.append(tf_idf_score)
    tfidf_values.append(tfidf_sentences)
```

```python
tf_idf_model = np.asarray(tfidf_values)
```

```python
tf_idf_model = np.transpose(tf_idf_model)
```

```python
from collections import Counter
import random
import numpy as np

threshold = 1e-5
word_counts = Counter(new_words)
total_count = len(new_words)
freqs = {word: count/total_count for word, count in word_counts.items()}
p_drop = {word: 1 - np.sqrt(threshold/freqs[word]) for word in word_counts}
train_words = [word for word in new_words if random.random() < (1 - p_drop[word])]
print(train_words)
```

```python
#Word embedding
# Python program to generate word vectors using Word2Vec
  
# importing all necessary modules
from nltk.tokenize import sent_tokenize, word_tokenize
import warnings
  
warnings.filterwarnings(action = 'ignore')
  
import gensim
from gensim.models import Word2Vec
  
#  Reads ‘alice.txt’ file
s = corpus
  
# Replaces escape character with space
f = s.replace("\n", " ")
  
data = []
  
# iterate through each sentence in the file
for i in sent_tokenize(f):
    temp = []
      
    # tokenize the sentence into words
    for j in word_tokenize(i):
        temp.append(j.lower())
  
    data.append(temp)
  
# Create CBOW model
model1 = gensim.models.Word2Vec(data, min_count = 1, 
                              vector_size = 100, window = 5)
  
# Print results
print("Cosine similarity between 'vaccines' " + 
               "and 'vaccinated' - CBOW : ",
    model1.wv.similarity('vaccines', 'vaccinated'))
      
print("Cosine similarity between 'vaccines' " +
                 "and 'booster' - CBOW : ",
      model1.wv.similarity('vaccines', 'booster'))
  
# Create Skip Gram model
model2 = gensim.models.Word2Vec(data, min_count = 1, vector_size = 100,
                                             window = 5, sg = 1)
  
# Print results
print("Cosine similarity between 'vaccines' " +
          "and 'vaccinated' - Skip Gram : ",
    model2.wv.similarity('vaccines', 'vaccinated'))
      
print("Cosine similarity between 'vaccines' " +
            "and 'booster' - Skip Gram : ",
      model2.wv.similarity('vaccines', 'booster'))
```
